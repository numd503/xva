```python
import numpy as np  
import pandas as pd
import datetime
from matplotlib import pyplot as plt
import seaborn as sns;
CHART_DIMENTIONS_SINGLE_CHART   = (20,7)
```

```python
all_by_notional = pd.read_excel("data/graphs.xlsx", sheet_name=0)
all_by_notional['TIME_PERIOD'] = pd.to_datetime(all_by_notional['TIME_PERIOD'])
all_by_notional = all_by_notional.set_index('TIME_PERIOD')
```

```python
(all_by_notional/ 1_000_000).plot(figsize=(10, 5))
plt.title('Notional Outstanding')
plt.ylabel('Trillion USD')
plt.xlabel('Year');
```
